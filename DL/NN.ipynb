# 神经网络向量化实现

# 激活函数（激活函数常用g表示，激活值用a表示，n表示特征数）：
1. Sigmoid 常用在输出层用于二元分类 (0,1)
2. Tanh   效果总是比sigmoid好 有类似数据真心化的效果，让下一层的学习更方便(-1,1)
3. ReLU Rectified Linear Unit，上述函数有个缺点，有梯度消失的情况，ReLU就不会，而且会快很多，热门选择
# 为什么需要非线性激活函数
如果没有非线性的激活函数，那再深的神经网络在做的也是线性的组合变换
# 激活函数的导数
g prime of z is a shorthand for the dg(z)/dz  
- Sigmoid: g'(z)=a(1-a)
- Tanh: g'(z)=1-a^2
- ReLU: 略
# 梯度下降法
偏导数衡量了多变量函数在某一个特定方向上的变化率，而梯度则是这些偏导数组成的向量

# 随机初始化
如果初始化权重为0，会发生完全对称现象
我们需要不同的神经元，需要随机初始化w，但b的话0即可
$w^[1]=np.random.randn((2,2))*0.01 $ 初始化成非常小的随机数,如果w太大，则使用Tanh和Sigmoid的时候梯度小，更新慢
$b^[1]=np.random.zeros((2,1))$
